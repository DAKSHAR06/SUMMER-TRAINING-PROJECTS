```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target  # 0 = malignant, 1 = benign

df.to_csv("breast_cancer.csv", index=False)
df.head()
df["target"].value_counts()
```

```python
import numpy as np
import pandas as pd
import keras
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
import pickle
```

```python
df = pd.read_csv("breast_cancer.csv")
df["target"].value_counts()
```

```python
df.tail()
```

```python
X = df.drop(columns=["target"])
y = df["target"]
```

```python
feature_names = X.columns.tolist()  # save for UI labels

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
train_X, test_X, train_y, test_y = train_test_split(X_scaled, y, test_size=0.3, random_state=123)
```

```python
model = Sequential()
model.add(Dense(units=30, kernel_initializer='uniform', activation='relu', input_dim=30))
model.add(Dense(units=16, kernel_initializer='uniform', activation='relu'))
model.add(Dense(units=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(units=4, kernel_initializer='uniform', activation='relu'))
model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(
    train_X,
    train_y,
    epochs=20,
    batch_size=32,
    validation_split=0.1
)
```

```python
model.save("ann_classifier.keras")
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
with open("feature_names.pkl", "wb") as f:
    pickle.dump(feature_names, f)
print("sa")
```
